```python
%matplotlib inline
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('..\Files\myself.jpg', cv.IMREAD_GRAYSCALE)

img0 = cv.imread('..\Files\myself.jpg', cv.IMREAD_GRAYSCALE) # Image is an array of numbers in the range [0,255]
img1 = img0 + 255 # Adding 50 to each of the pixel. Overflows can happen. value%256 (the modulo) will be assigned to the pixel in such a case.
img2 = cv.add(img0, 255) # No overflow. If the value exceeds 255. Value will be set to 255.(modulo is not considered)

fig, ax = plt.subplots(nrows = 1, ncols = 3, sharey=True)

ax[0].imshow(img0, cmap ='gray', vmin = 0, vmax = 255)
ax[0].set_title('img0')
ax[1].imshow(img1, cmap ='gray', vmin = 0, vmax = 255)
ax[1].set_title('img0 + 70')
ax[2].imshow(img2, cmap ='gray', vmin = 0, vmax = 255)
ax[2].set_title('cv.add(img0, 70)')
plt.show()
```

```python
%matplotlib inline
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('..\Files\myself.jpg', cv.IMREAD_GRAYSCALE)

# Transformation
transform = np.arange(0,256).astype('uint8')

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_title('Transformation')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')

fig, ax = plt.subplots(1,3)
ax[0].imshow(img, cmap = 'gray', vmin = 0, vmax = 255)
ax[0].set_title('Original')

img2 = cv.LUT(img, transform)
ax[1].imshow(img2, cmap = 'gray', vmin = 0, vmax = 255)
ax[1].set_title('Transformed')

img3 = img - img2
ax[2].imshow(img3, cmap = 'gray', vmin = 0, vmax = 255)
ax[2].set_title('Subtracted')

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

#img = cv.imread('anto.jpg')
imgo=cv.imread('anto.jpg', cv.IMREAD_COLOR)
img=cv.cvtColor(imgo, cv.COLOR_BGR2RGB)
a, b = [100,200]

tr1 = np.linspace(0,a,100).astype('uint8')
tr2 = np.linspace(a+1,b,56).astype('uint8')
tr3 = np.linspace(b+1,255,100).astype('uint8')

transf = np.concatenate((tr1, tr2), axis = 0)
transf = np.concatenate((transf, tr3), axis = 0)

fig, ax = plt.subplots()
ax.plot(transf)
ax.set_title('transformation')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()

fig, ax = plt.subplots(1,2)
ax[0].imshow(img, cmap ='gray', vmin = 0, vmax = 255)
ax[0].set_title('Original')

img2 = cv.LUT(img, transf)
ax[1].imshow(img2, cmap = 'gray', vmin = 0, vmax = 255)
ax[1].set_title('Transformed')
plt.show()
```

```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("anto.jpg", cv.IMREAD_COLOR)
channels = ['b', 'g', 'r'] # since cv reads img as BGR channel order
for i, col in enumerate(channels):
    histogram = cv.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histogram, color = col)
    plt.xlim([0, 255])
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('myself.jpg' ,cv.IMREAD_COLOR)
cv.namedWindow("Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Image", img)
cv.waitKey(0)


gamma = 0.5
table = np.array([(i/255.0)**(gamma)*255.0 for i in np.arange(0,256)]).astype('uint8')
img_gamma = cv.LUT(img, table)



cv.namedWindow("Gamma Corrected", cv.WINDOW_AUTOSIZE)
cv.imshow("Gamma Corrected", img_gamma)
cv.waitKey(0)
cv.destroyAllWindows()
```
